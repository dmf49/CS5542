```
import pandas as pd
import matplotlib.pyplot as plot



#open the csv file
data_df = pd.read_csv('/content/drive/My Drive/CS5542/Data.csv')

#display the column names
print('Displaying the column names:')
for col in data_df.columns:
  print(col)

#plot an 1 dimensional histogram
data_df[['radius_mean']].plot(kind='hist')
plot.show()

#plot a histogram of radius_mean by diagnosis
data_df.groupby('diagnosis')['radius_mean'].nunique().plot(kind='bar')
plot.show()

#create a scatter plot showing correlation between radius_mean vs texture_mean
data_df.plot.scatter(x='radius_mean', y='texture_mean', title= "radius_mean vs texture_mean");

plot.show();

#create a 3D plot to show correlation betweeb radius_mean, texture_mean and area_mean
plotThreeDimensional = plot.figure().gca(projection='3d')
plotThreeDimensional.scatter(data_df['radius_mean'], data_df['texture_mean'], data_df['area_mean'])
plotThreeDimensional.set_xlabel('radius_mean')
plotThreeDimensional.set_ylabel('texture_mean')
plotThreeDimensional.set_zlabel('area_mean')

plot.show()
```
